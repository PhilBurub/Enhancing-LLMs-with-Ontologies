```python
import requests
from bs4 import BeautifulSoup

URL = "https://web.archive.org/web/20240128215802/https://dief.tools.dbpedia.org/server/ontology/classes/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.findAll("a", href=False)
print(results)
```

```python
ul = soup.body.ul
```

```python
def dictify(ul):
    result = {}
    for li in ul.find_all("li", recursive=False):
        key = next(li.stripped_strings)
        ul = li.find("ul")
        if ul:
            result[key] = dictify(ul)
        else:
            result[key] = None
    return result

dictified = dictify(ul)
print(dictified)
```

```python
from benedict import benedict

d = benedict(dictified)

def traverse_item(d, key, value):
   for child in list(value):
      #f.write(str(child + ' is a subclass of ' + key))
      print(child + ' is a subclass of ' + key)
   #print('key: {} - value: {}'.format(key, list(value)))

d.traverse(traverse_item)
```
